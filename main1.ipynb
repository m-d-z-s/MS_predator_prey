```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.1
b = 0.02
c = 0.1
e = 0.02

init_condition = [c/e, a/b]
init_condition_2 = [c/e+0.1, a/b + 0.1]
init_condition_3 = [c/e+2, a/b + 1]

t = np.linspace(0, 500, 1000)
```

```python
def dpredator_dt(predator, prey):
    return -c * predator + e * predator * prey
```

```python
def dprey_dt(predator, prey):
    return a * prey - b * predator * prey
```

```python
def runge_kutta4(predator, prey, dt):
    k1 = dt * dpredator_dt(predator, prey)
    l1 = dt * dprey_dt(predator, prey)
    k2 = dt * dpredator_dt(predator + k1/2, prey + l1/2)
    l2 = dt * dprey_dt(predator + k1/2, prey + l1/2)
    k3 = dt * dpredator_dt(predator + k2/2, prey + l2/2)
    l3 = dt * dprey_dt(predator + k2/2, prey + l2/2)
    k4 = dt * dpredator_dt(predator + k3, prey + l3)
    l4 = dt * dprey_dt(predator + k3, prey + l3)
    return predator + (k1 + 2*k2 + 2*k3 + k4)/6, prey + (l1 + 2*l2 + 2*l3 + l4)/6
```

```python
predator = [init_condition[0]]
prey = [init_condition[1]]
for i in range(len(t)-1):
    pred, pray = runge_kutta4(predator[i], prey[i], t[i+1]-t[i])
    predator.append(pred)
    prey.append(pray)
    
predator_2 = [init_condition_2[0]]
prey_2 = [init_condition_2[1]]
for i in range(len(t)-1):
    pred, pray = runge_kutta4(predator_2[i], prey_2[i], t[i+1]-t[i])
    predator_2.append(pred)
    prey_2.append(pray)

predator_3 = [init_condition_3[0]]
prey_3 = [init_condition_3[1]]
for i in range(len(t)-1):
    pred, pray = runge_kutta4(predator_3[i], prey_3[i], t[i+1]-t[i])
    predator_3.append(pred)
    prey_3.append(pray)
```

```python
plt.figure(figsize=(8, 6))

# График для первой ситуации
plt.plot(5.0, 5.0, 'ro', label='Situation 1')

# График для второй ситуации
plt.plot(prey_2, predator_2, label='Situation 2')

# График для третьей ситуации
plt.plot(prey_3, predator_3, label='Situation 3')

plt.xlabel('Prey')
plt.ylabel('Predator')
plt.title('Phase Portrait')
plt.legend()
plt.grid(True)
plt.show()
```
